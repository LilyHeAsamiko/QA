### 2. 运行 Jupyter Notebook

**【Windows 系统】**

- 方法一：在开始菜单运行 Jupyter Notebook
- 方法二：运行 cmd 打开命令窗口，输入`jupyter notebook`并回车
- 方法三：在开始菜单运行 Anaconda Prompt，输入`jupyter notebook`并回车

Windows 系统指定路径打开 Jupyter Notebook：

- 步骤一：指定文件夹（保存.ipynb 文件的文件夹）空白处按住 shift 键同时点击右键
- 步骤二：弹出菜单栏点击“在此处打开命令窗口”
- 步骤三：在弹出的命令窗口输入`jupyter notebook`并回车

**【MacOS 系统】**

- 方法一：打开终端（terminal）输入`jupyter notebook` 并回车（推荐）
- 方法二：从 Anaconda-Navigator 运行 Jupyter Notebook

MacOS 系统指定路径打开 Jupyter Notebook：

- 步骤一：选中指定文件夹（保存.ipynb 文件的文件夹）并点击右键
- 步骤二：弹出菜单栏点击“New Terminal at Folder”
- 步骤三：在弹出的终端输入`jupyter notebook`并回车

---
### 3. 设置 Jupyter Notebook 默认浏览器

**如运行 Jupyter Notebook 不能自动弹出浏览器，则进行以下操作。**

将 Chrome 设置为 Jupyter Notebook 默认浏览器：

- 步骤一：打开命令窗口，输入命令`jupyter notebook --generate-config`并回车
- 步骤二：记事本打开“jupyter_notebook_config.py”配置文件，添加以下代码：

`import webbrowser
webbrowser.register("chrome",None,webbrowser.GenericBrowser(u"Chrome所在路径"))
c.NotebookApp.browser = 'chrome'`
