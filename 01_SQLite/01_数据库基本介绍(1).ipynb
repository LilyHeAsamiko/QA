## 1. 数据库

### 1.1. 什么是数据库？

首先，数据库是什么？简单来说，数据库就是一个存储数据的电脑文件，比如大家所熟知的 Microsoft Excel、csv、json 文件，甚至是一个普通的纯文本文件，如果它的目的是存储数据，也可以称为数据库。**数据库（Database，简称 DB）** 的定义是“按照数据结构来组织、存储和管理数据的仓库”，是一个长期存储在计算机内的、有组织的、可共享的、统一管理的大量数据的集合。

**数据库管理系统（Database Management System，简称 DBMS）** 是一种操纵和管理数据库的大型软件，用于建立、使用和维护数据库。它对数据库进行统一的管理和控制，以保证数据库的安全性和完整性。用户通过 DBMS 访问数据库中的数据，数据库管理员也通过 DBMS 进行数据库的维护工作。它可以支持多个应用程序和用户用不同的方法在同时或不同时刻去建立、修改和询问数据库。

### 1.2. 为什么使用数据库？

- 海量数据存储
- 方便管理和查找数据
- 数据得以永久保存
- 避免重复获取数据
- 多人共享数据
- ……

### 1.3. 主流数据库介绍

<font size=3>**关系型数据库**</font>

关系型数据库是指采用了关系模型来组织数据的数据库。关系模型可以简单理解为二维表格模型，而一个关系型数据库就是由二维表及其之间的关系组成的一个数据组织。虽然关系型数据库有很多，但是大多数都遵循 SQL（结构化查询语言，Structured Query Language）标准。常见的操作有查询、新增、更新、删除、排序等。关系型数据库对于结构化数据的处理更合适。

**关系数据库管理系统（Relational Database Management System：RDBMS）** 是指包括相互联系的逻辑组织和存取这些数据的一套程序。关系数据库管理系统就是管理关系数据库，并将数据逻辑组织的系统。

主流的关系型数据库管理系统：



**Oracle**

Oracle Database，又名 Oracle RDBMS，或简称 Oracle。是甲骨文公司的一款关系数据库管理系统。它是在数据库领域一直处于领先地位的产品。可以说 Oracle 数据库系统是目前世界上最流行的关系数据库管理系统，系统可移植性好、使用方便、功能强，适用于各类大、中、小微机环境。它是一种高效率的、可靠性好的、适应高吞吐量的数据库方案。

**MySQL**

MySQL 是一种开放源代码的关系型数据库管理系统（RDBMS），使用结构化查询语言（SQL）进行数据库管理。MySQL 是开放源代码的，因此任何人都可以在 General Public License 的许可下下载并根据个性化的需要对其进行修改。MySQL 因为其速度、可靠性和适应性而备受关注。大多数人都认为在不需要事务化处理的情况下，MySQL 是管理内容最好的选择。

**Microsoft SQL Server**

SQL Server 是 Microsoft 公司推出的关系型数据库管理系统，具有使用方便、可伸缩性好、与相关软件集成程度高等优点，可跨越从运行 Microsoft Windows 98 的膝上型电脑到运行 Microsoft Windows 2012 的大型多处理器的服务器等多种平台使用。

**PostgreSQL**

PostgreSQL 是以加州大学伯克利分校计算机系开发的 POSTGRES 为基础的对象关系型数据库管理系统（ORDBMS）。POSTGRES 领先的许多概念在很久以后才出现在一些商业数据库系统中。因为自由宽大的许可证，任何人都可以以任何目的免费使用、修改和分发 PostgreSQL，不管是私用、商用还是学术研究目的。

**IBM DB2**

IBM DB2 是美国IBM公司开发的一套关系型数据库管理系统，它主要的运行环境为UNIX（包括IBM自家的AIX）、Linux、IBM i（旧称OS/400）、z/OS，以及Windows服务器版本。DB2主要应用于大型应用系统，具有较好的可伸缩性，可支持从大型机到单用户环境，应用于所有常见的服务器操作系统平台下。 

**SQLite**

我们今天要使用的 SQLite 数据库是一款非常小巧的嵌入式开源数据库软件，虽然它的功能没有以上大型数据库那么全面，但它的使用非常简单方便，也不需要复杂的安装，因而非常适合初学者。

<font size=3>**非关系型数据库**</font>

随着近些年技术方向的不断拓展，大量的 NoSql 数据库管理系统如 MongoDB、Redis、Memcache 出于简化数据库结构、避免冗余、影响性能的表连接、摒弃复杂分布式的目的被设计。NoSQL 数据库适合追求速度和可扩展性、业务多变的应用场景。对于非结构化数据的处理更合适，如文章、评论，这些数据通常只用于模糊处理，并不需要像结构化数据一样进行精确查询，而且这类数据的数据规模往往是海量的，数据规模的增长往往也是不可能预期的，而 NoSQL 数据库的扩展能力几乎也是无限的，所以 NoSQL 数据库可以很好的满足这一类数据的存储。NoSQL 数据库利用 key-value 可以大量的获取非结构化数据，并且数据的获取效率很高，但用它查询结构化数据效果就比较差。

## 2. SQLite 数据库

### 2.1. SQLite 数据库基本介绍

SQLite 是一个进程内的库，实现了自给自足的、无服务器的、零配置的、事务性的 SQL 数据库引擎。它是一个零配置的数据库，这意味着与其他数据库不一样，您不需要在系统中配置。SQLite 应用广泛，增长迅速，作为后端数据库，可以搭配 Python 建网站，或者制作有数据存储需求的工具，比如 Android Apps。

**优点：**
- 轻量级，占用内存很小，大约250KB（支持2TB数据存储）
- 无服务器，零配置，无需安装（MySQL需要安装，设置用户名密码等等）
- 操作快，节省内存（只将需要的数据放到内存）

**缺点：**
- 不能承受高并发访问
- 功能较少
- 性能不是最优

**发展历史：**
- 2000 -- D. Richard Hipp 设计 SQLite 是为了不需要管理即可操作程序
- 2000 -- 在八月，SQLite1.0 发布 GNU 数据库管理器（GNU Database Manager）
- 2011 -- Hipp 宣布，向 SQLite DB 添加 UNQl 接口，开发 UNQLite（面向文档的数据库）

### 2.3. 认识 SQLite 数据库基本结构

在 **“数据库结构”** 界面，可以看到整个数据库的结构，包括所有表格的名称，以及各表格的所有列名。



在 **“浏览数据”** 界面，可以浏览表中数据。



数据库表格的一般形式为：行（row)、列（column）、字段（field），每行表示一条“记录”。

SQLite 支持五种存储类：
- NULL
- TEXT
- INTEGER
- REAL
- BLOB

SQLite 没有单独的 Boolean 存储类。相反，布尔值被存储为整数 0（false）和 1（true）。

SQLite 没有一个单独的用于存储日期和/或时间的存储类，但 SQLite 能够把日期和时间存储为 TEXT、REAL 或 INTEGER 值。

## 3. sqlite3 模块

尽管使用 Browser 我们就可以对数据库进行阅读和修改，但在 Python 项目过程中，我们往往需要使用 Python 语言与数据库进行交互，从而进行数据录入、更新、读取和分析。好消息是，Python 标准库中的 sqlite3 模块可以完成与 SQLite 数据库的交互！这就意味着你不需要下载任何软件，甚至不需要 pip install 就可以使用。

### 3.1. 导入模块

```python
import sqlite3
```

### 3.2. 创建数据库连接

```python
conn = sqlite3.connect('Northwind_small.sqlite')      # 如果数据库文件不存在则新建一个数据库
```

### 3.3. 创建光标

```python
cur = conn.cursor()
```

### 3.4. 使用光标执行命令

```python
cur.execute("SELECT * FROM Product")
# 光标每次只能执行一条命令，所以末尾不加分号也可以
```

```python
# 也可以将 sql 语句先赋值给变量
sql = "SELECT * FROM Product"
cur.execute(sql)
```

### 3.5. 获取数据或保存修改

```python
# 获取光标中存储的数据
for row in cur:
    print(row)
```

```python
for row in cur[:5]:         # 报错，cursor 对象不是列表
    print(row)
```

```python
# 获取光标中存储的下一条记录
cur.execute(sql)
cur.fetchone()
```

```python
# 获取光标中多条记录，返回一个列表
cur.fetchmany(5)
```

```python
for row in cur.fetchmany(5):          
    print(row)
```

```python
# 获取光标中剩余所有记录
cur.fetchall()      
```

```python
# 保存修改（增删改）
conn.commit()
```

### 3.6. 关闭连接

```python
conn.close()
```

**注意**
- 在最开始创建连接
- 中间每完成一项工作可以保存
- 最后关闭连接，不要每一步都关闭连接再打开

### 练习

```python
# 创建连接和光标
conn = sqlite3.connect('Northwind_small.sqlite') 
cur = conn.cursor()
```

```python
# 需求 1：打印出英国的所有客户的公司名称
# 提示：使用 WHERE 进行限制
cur.execute("SELECT CompanyName FROM Customer WHERE Country='UK'")
cur.fetchall()
```

```python
cur.execute("SELECT CompanyName FROM Customer WHERE Country='UK'")
for i in cur:
    print(i[0])             # 使用元组索引，只打印公司名称
```

```python
# 需求 2：打印出美国所有员工姓名
cur.execute("SELECT LastName,FirstName FROM Employee WHERE Country='USA'")
for last, first in cur:
    print(last, first)      # 使用元组解包赋值
```

```python
# 中间每完成一步操作可以进行保存
conn.commit()
```

```python
# 执行完所有操作之后，关闭连接
conn.close()
```
